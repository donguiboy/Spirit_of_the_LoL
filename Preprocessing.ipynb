```python
import requests
import pandas as pd

from dotenv import load_dotenv
import os

```

```python
# Cargar el archivo .env
load_dotenv()

# Acceder al valor de una variable
api_key = os.getenv("API_KEY")
```

```python
import requests

url = f'https://americas.api.riotgames.com/lol/match/v5/matches/LA2_1294816603?api_key={api_key}'
# Data de partidas, no por tiempo

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    
    # Process the data as needed
    print(data)
else:
    print('Error:', response.status_code)
```

```python
df = pd.DataFrame(data)
df
```

```python
data_info = data["info"]
data_info
```

```python
data_participants_detalle = data_info["participants"] #Devuelve diccionario con el detalle para cada participante (del 0 al 9)
data_participants_detalle
```

```python
# Create an empty list to store the DataFrames
participant_dataframes = []

# Create a DataFrame for each participant
for participant_info in data_participants_detalle:
    participant_df = pd.DataFrame([participant_info])
    participant_dataframes.append(participant_df)

# Concatenate all participant DataFrames into a single DataFrame
participant_df = pd.concat(participant_dataframes)

# Reset the index of the resulting DataFrame
participant_df= participant_df.reset_index(drop=True)

# Transpose the DataFrame to swap rows and columns
#participant_df = participant_df.transpose()

print(participant_df)
```

```python
# Lee el archivo de Excel
excel_info = pd.read_excel('participant.xlsx', sheet_name='Sheet1')

# Aplica la condición y obtén la columna como lista ----> seleccionar las columnas del excel que establecimos como "drop"
drop_columns = excel_info.loc[excel_info['Operación'] == 'drop']['feature'].tolist()
```

```python
#DF sin las features seleccionadas previamente como "drop"
reduced_df = participant_df.drop(drop_columns, axis = 1)
reduced_df
```
